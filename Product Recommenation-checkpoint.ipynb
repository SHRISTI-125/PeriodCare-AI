```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import precision_score
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix

# Load dataset
df = pd.read_csv("menstrual_product_recommendation.csv")

# Encode categorical variables
label_encoders = {}
categorical_columns = ["Age_group", "Skin_Sensitivity", "Flow_Level", "Fragrance", "Lifestyle", "Pain_Level", "Eco-Friendly"]

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Store encoders for future decoding

# Scale the features (MinMaxScaler)
scaler = MinMaxScaler()
df_features = df[categorical_columns]
df_features_matrix = df_features.values
df_features_matrix_scaled = scaler.fit_transform(df_features_matrix)

# Split data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
train_features = train_df[categorical_columns]
test_features = test_df[categorical_columns]

# 🔹 Function to recommend products based on multiple features
def recommend_products(user_input, n_neighbors=5, metric='cosine'):
    """
    Recommend products based on user attributes using KNN.

    Parameters:
    - user_input: List of categorical feature values (Age, Sensitivity, Flow, etc.)
    - n_neighbors: Number of neighbors to search for
    - metric: Distance metric to use for KNN (e.g., 'cosine', 'euclidean')

    Returns:
    - List of recommended products
    """
    expected_features = len(categorical_columns)  # Ensure correct feature count

    if len(user_input) != expected_features:
        return f"Error: Input must have exactly {expected_features} features, but got {len(user_input)}."

    # Convert user input to NumPy array (remove feature names)
    user_array = np.array(user_input).reshape(1, -1)
    user_array_scaled = scaler.transform(user_array)  # Scale user input
    
    # Train KNN Model with specified n_neighbors and metric
    model_knn = NearestNeighbors(n_neighbors=n_neighbors, metric=metric)
    model_knn.fit(df_features_matrix_scaled)  # Fit using scaled features

    # Find closest match
    distances, indices = model_knn.kneighbors(user_array_scaled, n_neighbors=n_neighbors + 1)

    # Recommend products from closest users (ignoring the first one, as it's the user themselves)
    recommended_products = df.iloc[indices.flatten()]['Product'][1:].tolist()
    return recommended_products

# 🔹 Get recommendations for the test set
def get_recommendations_for_test_set(test_df, n_neighbors=5, metric='cosine'):
    recommendations = []

    for index, row in test_df.iterrows():
        user_input = row[categorical_columns].values
        recommended_products = recommend_products(user_input, n_neighbors, metric)
        recommendations.append(recommended_products)

    return recommendations

# 🔹 Calculate Precision at K
def precision_at_k(recommendations, actual_products, k=3):
    precision_scores = []
    for recommended, actual in zip(recommendations, actual_products):
        relevant_recommended = [1 if product in actual else 0 for product in recommended]
        precision_scores.append(sum(relevant_recommended) / k)
    return np.mean(precision_scores)

# Get recommendations for test set
recommendations = get_recommendations_for_test_set(test_df, n_neighbors=5, metric='cosine')

# Calculate Precision at K (assuming 'Actual_Product' contains the actual product for comparison)
actual_products = test_df['Product'].values.tolist()  # Ensure 'Actual_Product' column exists
precision = precision_at_k(recommendations, actual_products, k=3)
print(f"Precision at K: {precision:.2f}")

# 🔹 Calculate overall accuracy of recommendations
def calculate_accuracy(test_df, recommendations):
    correct_count = 0
    for i, recommended in enumerate(recommendations):
        actual_product = test_df.iloc[i]['Product']
        if actual_product in recommended:
            correct_count += 1
    
    accuracy = correct_count / len(test_df)
    return accuracy

# Calculate accuracy
accuracy = calculate_accuracy(test_df, recommendations)
print(f"Recommendation Accuracy: {accuracy * 100:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD

# Load dataset
df = pd.read_csv("menstrual_product_recommendation.csv")

# Encode categorical variables
label_encoders = {}
categorical_columns = ["Age_group", "Skin_Sensitivity", "Flow_Level", "Fragrance", "Lifestyle", "Pain_Level", "Eco-Friendly"]

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Store encoders for future decoding

# Scale the features (MinMaxScaler)
scaler = MinMaxScaler()
df_features = df[categorical_columns]
df_features_matrix = df_features.values
df_features_matrix_scaled = scaler.fit_transform(df_features_matrix)

# Split data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
train_features = train_df[categorical_columns]
test_features = test_df[categorical_columns]

# 🔹 Function to recommend products based on multiple features
def recommend_products(user_input, n_neighbors=5, metric='cosine'):
    """
    Recommend products based on user attributes using KNN.

    Parameters:
    - user_input: List of categorical feature values (Age, Sensitivity, Flow, etc.)
    - n_neighbors: Number of neighbors to search for
    - metric: Distance metric to use for KNN (e.g., 'cosine', 'euclidean')

    Returns:
    - List of recommended products
    """
    expected_features = len(categorical_columns)  # Ensure correct feature count

    if len(user_input) != expected_features:
        return f"Error: Input must have exactly {expected_features} features, but got {len(user_input)}."

    # Convert user input to NumPy array (remove feature names)
    user_array = np.array(user_input).reshape(1, -1)
    user_array_scaled = scaler.transform(user_array)  # Scale user input
    
    # Train KNN Model with specified n_neighbors and metric
    model_knn = NearestNeighbors(n_neighbors=n_neighbors, metric=metric)
    model_knn.fit(df_features_matrix_scaled)  # Fit using scaled features

    # Find closest match
    distances, indices = model_knn.kneighbors(user_array_scaled, n_neighbors=n_neighbors + 1)

    # Recommend products from closest users (ignoring the first one, as it's the user themselves)
    recommended_products = df.iloc[indices.flatten()]['Product'][1:].tolist()
    return recommended_products

# 🔹 Matrix Factorization with SVD for further improvement
def svd_recommendation(train_features, n_components=5):
    """
    Apply Singular Value Decomposition (SVD) to reduce the dimensionality and get better recommendations.
    
    Parameters:
    - train_features: The training feature matrix
    - n_components: Number of components for dimensionality reduction

    Returns:
    - Reduced matrix after SVD
    """
    svd = TruncatedSVD(n_components=n_components)
    reduced_matrix = svd.fit_transform(train_features)
    return reduced_matrix

# 🔹 Get recommendations for the test set
def get_recommendations_for_test_set(test_df, n_neighbors=5, metric='cosine'):
    recommendations = []

    for index, row in test_df.iterrows():
        user_input = row[categorical_columns].values
        recommended_products = recommend_products(user_input, n_neighbors, metric)
        recommendations.append(recommended_products)

    return recommendations

# 🔹 Calculate Precision, Recall, and F1-Score at K
def precision_recall_f1_at_k(recommendations, actual_products, k=3):
    precision_scores = []
    recall_scores = []
    f1_scores = []
    
    for recommended, actual in zip(recommendations, actual_products):
        relevant_recommended = [1 if product in actual else 0 for product in recommended]
        tp = sum(relevant_recommended)
        fp = k - tp
        fn = len(actual) - tp
        
        # Calculate Precision, Recall, and F1
        precision = tp / (tp + fp) if tp + fp > 0 else 0
        recall = tp / (tp + fn) if tp + fn > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
        
        precision_scores.append(precision)
        recall_scores.append(recall)
        f1_scores.append(f1)
    
    avg_precision = np.mean(precision_scores)
    avg_recall = np.mean(recall_scores)
    avg_f1 = np.mean(f1_scores)
    
    return avg_precision, avg_recall, avg_f1

# Get recommendations for test set
recommendations = get_recommendations_for_test_set(test_df, n_neighbors=5, metric='cosine')

# Calculate Precision, Recall, and F1-Score at K
actual_products = test_df['Product'].values.tolist()  # Ensure 'Actual_Product' column exists
precision, recall, f1 = precision_recall_f1_at_k(recommendations, actual_products, k=3)
print(f"Precision at K: {precision:.2f}")
print(f"Recall at K: {recall:.2f}")
print(f"F1-Score at K: {f1:.2f}")

# 🔹 Calculate overall accuracy of recommendations
def calculate_accuracy(test_df, recommendations):
    correct_count = 0
    for i, recommended in enumerate(recommendations):
        actual_product = test_df.iloc[i]['Product']
        if actual_product in recommended:
            correct_count += 1
    
    accuracy = correct_count / len(test_df)
    return accuracy

# Calculate accuracy
accuracy = calculate_accuracy(test_df, recommendations)
print(f"Recommendation Accuracy: {accuracy * 100:.2f}%")

# 🔹 Apply SVD for dimensionality reduction
reduced_matrix = svd_recommendation(df_features_matrix_scaled, n_components=5)

# Print the shape of the reduced matrix
print(f"Reduced matrix shape after SVD: {reduced_matrix.shape}")
```

```python
def precision_recall_f1_at_k(recommendations, actual_products, k=5):
    precision_scores = []
    recall_scores = []
    f1_scores = []
    
    for recommended, actual in zip(recommendations, actual_products):
        if isinstance(actual, list):  # Ensure actual products are a list
            actual_set = set(actual)
        else:
            actual_set = {actual}
        
        recommended_set = set(recommended[:k])
        
        tp = len(recommended_set & actual_set)  # True Positives
        fp = len(recommended_set - actual_set)  # False Positives
        fn = len(actual_set - recommended_set)  # False Negatives
        
        precision = tp / (tp + fp) if tp + fp > 0 else 0
        recall = tp / (tp + fn) if tp + fn > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
        
        precision_scores.append(precision)
        recall_scores.append(recall)
        f1_scores.append(f1)
    
    return np.mean(precision_scores), np.mean(recall_scores), np.mean(f1_scores)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Train SVD on scaled features
svd = TruncatedSVD(n_components=5)
reduced_matrix = svd.fit_transform(df_features_matrix_scaled)

# Build cosine similarity matrix
similarity_matrix = cosine_similarity(reduced_matrix)

# 🔹 Improved Recommendation Function
def recommend_products_svd(user_input, n_neighbors=5):
    """
    Recommend products using SVD-based nearest neighbors.
    
    Parameters:
    - user_input: List of categorical feature values.
    - n_neighbors: Number of neighbors to search for.

    Returns:
    - List of recommended products.
    """
    # Convert user input to NumPy array
    user_array = np.array(user_input).reshape(1, -1)
    user_array_scaled = scaler.transform(user_array)
    
    # Transform user input using SVD
    user_svd = svd.transform(user_array_scaled)
    
    # Compute similarity with all products
    similarities = cosine_similarity(user_svd, reduced_matrix)
    similar_indices = np.argsort(similarities[0])[-n_neighbors-1:-1]  # Top N excluding self
    
    recommended_products = df.iloc[similar_indices]['Product'].tolist()
    return recommended_products
```

```python
def get_hybrid_recommendations(test_df, n_neighbors=5):
    recommendations = []
    
    for _, row in test_df.iterrows():
        user_input = row[categorical_columns].values
        knn_recs = recommend_products(user_input, n_neighbors)
        svd_recs = recommend_products_svd(user_input, n_neighbors)
        
        # Combine & remove duplicates
        combined_recommendations = list(set(knn_recs + svd_recs))[:n_neighbors]
        recommendations.append(combined_recommendations)
    
    return recommendations
```

```python
def get_weighted_hybrid_recommendations(test_df, n_neighbors=5, svd_weight=0.7):
    recommendations = []
    
    for _, row in test_df.iterrows():
        user_input = row[categorical_columns].values
        
        knn_recs = recommend_products(user_input, n_neighbors)
        svd_recs = recommend_products_svd(user_input, n_neighbors)
        
        # Weighting SVD recommendations higher than KNN
        combined_recommendations = list(set(svd_recs[:int(n_neighbors * svd_weight)] + knn_recs[:int(n_neighbors * (1 - svd_weight))]))
        
        recommendations.append(combined_recommendations[:n_neighbors])  # Keep only top N
    
    return recommendations
```

```python
def recommend_products_filtered(user_input, n_neighbors=5, metric='cosine', similarity_threshold=0.5):
    user_array = np.array(user_input).reshape(1, -1)
    user_array_scaled = scaler.transform(user_array)
    
    model_knn = NearestNeighbors(n_neighbors=n_neighbors, metric=metric)
    model_knn.fit(df_features_matrix_scaled)
    
    distances, indices = model_knn.kneighbors(user_array_scaled, n_neighbors=n_neighbors + 1)
    
    recommended_products = []
    for idx, dist in zip(indices.flatten()[1:], distances.flatten()[1:]):  # Skip self
        if dist < similarity_threshold:
            recommended_products.append(df.iloc[idx]['Product'])
    
    return recommended_products
def get_ranked_hybrid_recommendations(test_df, n_neighbors=5):
    recommendations = []
    
    for _, row in test_df.iterrows():
        user_input = row[categorical_columns].values
        
        knn_recs = recommend_products_filtered(user_input, n_neighbors)
        svd_recs = recommend_products_svd(user_input, n_neighbors)
        
        # Assigning weights: Higher ranks get more points
        product_scores = {}
        
        for rank, product in enumerate(svd_recs):
            product_scores[product] = product_scores.get(product, 0) + (n_neighbors - rank)
        for rank, product in enumerate(knn_recs):
            product_scores[product] = product_scores.get(product, 0) + (n_neighbors - rank) / 2  # Lower weight for KNN
        
        # Sort recommendations based on scores
        ranked_recommendations = sorted(product_scores, key=product_scores.get, reverse=True)
        
        recommendations.append(ranked_recommendations[:n_neighbors])
    
    return recommendations
```

```python
def get_confidence_ranked_recommendations(test_df, n_neighbors=5):
    recommendations = []
    
    for _, row in test_df.iterrows():
        user_input = row[categorical_columns].values
        
        knn_recs = recommend_products_filtered(user_input, n_neighbors)
        svd_recs = recommend_products_svd(user_input, n_neighbors)
        
        product_scores = {}

        for rank, product in enumerate(svd_recs):
            product_scores[product] = product_scores.get(product, 0) + (n_neighbors - rank) * 1.2  # Higher weight for SVD
        
        for rank, product in enumerate(knn_recs):
            product_scores[product] = product_scores.get(product, 0) + (n_neighbors - rank) * 0.8  # Lower weight for KNN
        
        # If a product appears in both lists, boost its score
        for product in set(svd_recs) & set(knn_recs):
            product_scores[product] *= 1.5  # Boost confidence
        
        ranked_recommendations = sorted(product_scores, key=product_scores.get, reverse=True)
        
        recommendations.append(ranked_recommendations[:n_neighbors])
    
    return recommendations
# Generate confidence-based hybrid recommendations
confidence_hybrid_recommendations = get_confidence_ranked_recommendations(test_df, n_neighbors=5)

# Evaluate precision, recall, and F1-score
precision, recall, f1 = precision_recall_f1_at_k(confidence_hybrid_recommendations, actual_products, k=3)
print(f"Confidence Hybrid Model - Precision at K: {precision:.2f}")
print(f"Confidence Hybrid Model - Recall at K: {recall:.2f}")
print(f"Confidence Hybrid Model - F1-Score at K: {f1:.2f}")

# Calculate accuracy
accuracy = calculate_accuracy(test_df, confidence_hybrid_recommendations)
print(f"Confidence Hybrid Model - Recommendation Accuracy: {accuracy * 100:.2f}%")
```
