```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("excercise.csv")  
```

```python
label_encoders = {}
```

```python
for col in ["Age Group", "Flow Level", "Health Condition", "Pain Level"]:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[["Age Group", "Flow Level", "Health Condition", "Pain Level"]])
```

```python
k = 12  
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
print(df.groupby("Cluster")["Exercise Recommended"].value_counts())
```

```python
def predict_cluster(age, flow, health, pain):
    input_data = np.array([[age, flow, health, pain]])
    input_scaled = scaler.transform(input_data)
    cluster = kmeans.predict(input_scaled)[0]
    return cluster

new_user_cluster = predict_cluster(1, 2, 1, 0)  
print(f"New user's Cluster {new_user_cluster}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Age Group"], y=df["Flow Level"], hue=df["Cluster"], palette="plasma")
plt.title("K-Means Clustering on Menstrual Health Data")
plt.xlabel("Age Group")
plt.ylabel("Flow Level")
plt.legend(title="Cluster")
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)  
X_pca = pca.fit_transform(X_scaled)

explained_variance = pca.explained_variance_ratio_
print(f"Variance by PCA: {explained_variance}")
```

```python
wcss = []
K_range = range(2, 20)  
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_) 
```

```python
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o', linestyle='--')
plt.xlabel("Number K Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method After PCA")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_scores = []
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_pca)
    sil_scores.append(silhouette_score(X_pca, cluster_labels))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(K_range, sil_scores, marker='o', linestyle='--', color='r')
plt.xlabel("K Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score After PCA")
plt.show()

best_k = K_range[sil_scores.index(max(sil_scores))]
print(f"Optimal k: {best_k} && Silhouette Score: {max(sil_scores)}")
```
